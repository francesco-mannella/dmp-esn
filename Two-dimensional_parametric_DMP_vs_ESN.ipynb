# Comparing DMPs and ESNs

Here you find some tests comparing Stulp's parametric DMPs and ESNs in reproducing parametrized trajectories.

Brief summary of the two models:

* Parametric DMPs 
    * learning by optimizing a **forcing component**  whose dynamics adds to the dynamics a PD controller as in standard DMPs.  
    * The forcing component is a function of the canonical system as in standard DMPs.
    * The forcing component is also a function of further **given parameters**.
    * The forcing component is the weighted sum of the activation of a set multivariated gaussian basis functions spanning the space of the canonical system and the **parameters**.
    
<img src="pics/stulp-parametric.png" width=400 >    
    
* Parametric ESNs
    * a readout unit for each dimension of the trajectory to reproduce
    * the input to the ESN is an **inpulse** given at the starting time of the learning/reproduction of the trajectory.
    * the impulse is made by info about the  **parameters**. The **starting point** and the **end point** are considered as the other **parameters**.
    * learning involves the update of the weights from the units of the reservoir to the readout units through regression.
    * learning and reproduction of the trajectories is based on the activation of the readout units biased by the value of the end point. By this way we exploit the zero-attractor feature of ESN to obtain the stability of the end point.

<img src="pics/esn_discrete_2D_general.png" width=400 >
